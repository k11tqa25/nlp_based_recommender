# Retrieve Data

### We and our friends

```python
excel_df = pd.read_excel('comment 40 comedians.xlsx', index_col=0)
excel_df.head(1)
```

```python
excel_df.columns = ['Comedian','Leo','Kevin','Peter','Bangzhu']
```

```python
score_df = pd.DataFrame(excel_df[['Leo','Kevin','Peter','Bangzhu']].stack())
score_df = score_df.reset_index()
score_df.columns = ['Comedian_id','Name','Comment']
score_df.head()
```

```python
excel_df['Comedian_id'] = excel_df.index
score_df = score_df.merge(excel_df[['Comedian_id','Comedian']], on='Comedian_id')[['Comedian','Name','Comment']]
score_df.head()
```

```python
pol = lambda x: tb(x).sentiment.polarity
sub = lambda x: tb(x).sentiment.subjectivity
PorN = lambda x: x*0 if x==0 else (x*(1/x) if x>0 else x*(-1/x))
score_df['Polarity'] = score_df['Comment'].apply(pol)
score_df['Subjectivity'] = score_df['Comment'].apply(sub)
score_df['P/N'] = score_df['Polarity'].apply(PorN)
score_df = score_df.drop('Comment', axis=1)
score_df
```

```python
our_own_data = score_df.pivot_table(index=['Name'], columns=['Comedian'],values=['Subjectivity'], aggfunc='mean')
rename_columns = [c[1] for c in our_own_data.columns]
our_own_data.columns = rename_columns
our_own_data
```

```python
comedians_40 = excel_df.Comedian.unique()
```

## Compress rows (10 to 1)

```python
# First give each row a group number
ddd = df_UCM.copy()
ddd = ddd.reset_index()
ddd = ddd.rename(columns={'Name':'User_ID'})
ddd['User_ID'] = ddd.index%(int(len(ddd.index)/10))

# Group by the group number
ddd = ddd.groupby('User_ID').mean()

# Randomly assign the names to the users
np.random.seed(1)
ddd.index = np.random.choice(df_clean.Name, size=(ddd.shape[0])) 

# Append our own data
df_full = ddd.append(our_own_data)
df_full

df_full.to_pickle('Full_UCM.pkl')
df_full
```

## Split to train and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
stack = df_full.stack()
data = np.array([[v[0], v[1], stack.iloc[i]] for i, v in enumerate(stack.index)])
dataset = pd.DataFrame(data, columns=['Name', 'Comedian','Rating'])
dataset['Rating'] = dataset['Rating'].astype(float)
dataset
```

```python
# merge duplicated rows
dataset = dataset.groupby(['Name','Comedian'], as_index=False).mean()
dataset.head()
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(dataset[['Name','Comedian']], dataset['Rating'],stratify=dataset['Name'], test_size=0.2, random_state=1)
X_train = X_train.reset_index()[['Name','Comedian']]
X_test = X_test.reset_index()[['Name','Comedian']]
y_train = y_train.reset_index()['Rating'].astype(float)
y_test = y_test.reset_index()['Rating'].astype(float)

X_train.to_pickle('X_train.pkl')
X_test.to_pickle('X_test.pkl')
y_train.to_pickle('Y_train.pkl')
y_test.to_pickle('Y_test.pkl')
```
