```python
pol = lambda x: tb(x).sentiment.polarity
sub = lambda x: tb(x).sentiment.subjectivity

data_SA = pd.DataFrame()
data_SA['Title'] = df_webapage['Title']
data_SA['Name'] = df_webapage['Name']
data_SA['Comment'] = df_webapage['Comment']
data_SA['Polarity'] = df_webapage['Comment'].apply(pol)
data_SA['Subjectivity'] = df_webapage['Comment'].apply(sub)
data_SA
```

```python
PorN = lambda x: x*0 if x==0 else (x*(1/x) if x>0 else x*(-1/x))
data_SA['P/N'] = data_SA['Polarity'].apply(PorN)
data_SA
```

```python
data_SA = data_SA[data_SA.Subjectivity != 0]
```

## Compress rows (10 to 1)

```python
# First give each row a group number
ddd = df_UCM.copy()
ddd = ddd.reset_index()
ddd = ddd.rename(columns={'Name':'User_ID'})
ddd['User_ID'] = ddd.index%(int(len(ddd.index)/10))

# Group by the group number
ddd = ddd.groupby('User_ID').mean()
ddd.index = [f"User{i}" for i in ddd.index]
ddd.to_pickle('Full_UCM.pkl')
ddd
```

## Split to train and test dataset

```python
stack = ddd.stack()
data = np.array([[v[0], v[1]] for v in stack.index])
dataset = pd.DataFrame(data, columns=['Name', 'Comedian'])
dataset
```

```python
user_comedian = {}
for k, v in dataset.iterrows():
    user = v.Name
    comedian = v.Comedian
    if user not in user_comedian:
        user_comedian[user] = [comedian]
    else:
        user_comedian[user].append(comedian)

train_df = pd.DataFrame(columns=['Name', 'Comedian'])
test_df = pd.DataFrame(columns=['Name', 'Comedian'])
for k, v in user_comedian.items():
    n = len(v)
    if n < 6:
        test_size = 1
    elif n < 9:
        test_size = 2
    else:
        test_size = 3
    random.shuffle(v)
    for com in v[:(n-test_size)]:
        train_df = train_df.append({'Name':k, 'Comedian':com}, ignore_index=True)
    for com in v[-test_size:]:
        test_df = test_df.append({'Name':k, 'Comedian':com}, ignore_index=True)
```

```python
train_df.to_csv('training_set.csv')
train_df
```

```python
test_df.to_csv('test_set.csv')
test_df
```
